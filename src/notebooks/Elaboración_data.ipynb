## Categorías de Vino

Los distintos vinos se clasifican en diferentes categorías en función de la edad, es decir, el tiempo que ha permanecido envejeciendo tanto en barrica como en botella. Esta clasificación variará según se trate de una clase de vino u otra, siendo especialmente importante en el vino tinto. Además, hay que tener presente que cada D.O. establece sus características particulares; sin embargo, a efectos de este análisis nos guiaremos por una clasificación general.

#### Categorías de Vino Tinto
Atendiendo a lo anterior, distinguimos las siguientes categorías, que van de menor a mayor tiempo de envejecimiento:

            * Joven --> no pasan tiempo en barrica; el de la cosecha
            * Roble --> tiempo total entre barrica y botella 9 meses
            * Crianza --> tiempo total entre barrica y botella 24 meses
            * Reserva --> tiempo total entre barrica y botella 48 meses
            * Gran Reserva --> tiempo total entre barrica y botella 60 meses

#### Categorías de Vino Blanco y Rosado
En esta ocasión, los tiempos de envejecimiento difieren con respecto a la clase anterior, por lo que se tienen las siguientes categorías, pues no es tan común en estos vinos la crianza en barrica:

            * Reserva --> tiempo total 24 meses
            * Gran reserva --> tiempo total 36 meses
            
#### Categorías de Vino Espumoso
Por último, los vinos espumosos no atienden a las categorías mencionadas; sino que se clasifican por su método de elaboracion en:

            * Non-vintage --> elaboración con mezcla de añadas para conservar la calidad año tras año
            * Vintage



```python
######################### CATEGORIZACIÓN DE TINTOS ##############################
'''Para asignarles categoría a los all_vinos tintos vamos a definir 2 funciones:
        1ª) Se basará en el propio nombre del vino para determinar su categoría
        2ª) Si el nombre no menciona nada, recurriremos al año; de este modo:
                    * Joven --> 2021
                    * Roble --> 2020 y 2019
                    * Crianza --> 2018 y 2017
                    * Reserva --> 2016 y 2015
                    * Gran Reserva --> 2014 e inferiores
'''
def cat_nombre(x):
    if "gran reserva" in x.lower():
        return "Gran Reserva"

    elif "reserva" in x.lower():
        return "Reserva"

    elif "crianza" in x.lower():
        return "Crianza"

    elif "roble" in x.lower():
        return "Roble"

    elif "joven" in x.lower():
        return "Joven"

    return "Sin determinar"

tintos["Categoría"] = tintos["Vino"].apply(cat_nombre)

def cat_annio(y):
    if y == 2021:
        return "Joven"

    elif y == 2020 or y == 2019:
        return "Roble"

    elif y == 2018 or y==2017:
        return "Crianza"

    elif y==2016 or y==2015:
        return "Reserva"

    elif y<=2014:
        return "Gran Reserva"


tintos.loc[tintos[tintos["Categoría"] == "Sin determinar"].index,"Categoría"]= tintos[tintos["Categoría"] == "Sin determinar"]["Año"].apply(cat_annio)

tintos
```

```python
############################## CATEGORIZACIÓN DE BLANCOS Y ROSADOS ################################
'''Dado lo poco común que es la crianza de estos vinos en barrica,
    para clasificarlos sólo atenderemos al nombre; 
    por lo que una gran cantidad de vinos blancos y rosados permanecerán sin determinar'''

blancos_rosados = all_vinos[(all_vinos.Clase == "Blanco") | (all_vinos.Clase == "Rosado")]
blancos_rosados["Categoría"] = blancos_rosados["Vino"].apply(cat_nombre)
blancos_rosados
```

```python
############################## CATEGORIZACIÓN ESPUMOSOS ###########################
'''La clasificación en este caso es distinta, por lo que, de nuevo, sólo nos fijaremos en el 
    propio nombre del vino para designarlo como "Vintage" o "Non Vintage'''
    
espumosos = all_vinos[all_vinos.Clase == "Espumoso"]

import re
for i in espumosos[espumosos["Vino"].str.contains("vintage", flags=re.IGNORECASE)].index:
    espumosos.loc[i, "Categoría"] = "Vintage"

for i in espumosos[espumosos["Vino"].str.contains("n.v.", flags=re.IGNORECASE)].index:
    espumosos.loc[i, "Categoría"] = "Non Vintage"

for i in espumosos[espumosos["Categoría"].isna()].index:
    espumosos.loc[i, "Categoría"] = "Sin determinar"

espumosos
```

```python
vinos_cat = pd.concat([tintos,blancos_rosados,espumosos], ignore_index=True)

vinos_cat
```

## Denominación de Origen
Por último, vamos a determinar si la región de procedencia del vino tiene D.O. (denominación de origen), ya que así podremos analizar la influencia de este factor.

Además de la D.O, existen otras expresiones tradicionales como son los Vinos de Pago, los Vinos de Calidad, así como la Indicación Geográfica Protegida; sin embargo, tales denominaciones no se tendrán en cuenta en este análisis dado que gozan de escasa popularidad entre el público general y, por tanto, su influencia en la valoración de un vino por parte de cualquier persona ajena al mundo vitivinícola será ínfima.

```python
vinos_cat["Región"].unique()
```

```python
denominaciones_origen = ['Alicante','La Mancha','Rioja','Terra Alta', 'Cataluña', 'Cariñena', 'Jumilla',
                         'Navarra', 'Somontano', 'Valdepeñas', 'Utiel-Requena', 'Penedès','Campo de Borja',
                         'Montsant', 'Empordà', 'Ribera del Duero', 'Almansa', 'Toro','Yecla', 'Calatayud',
                         'Plà i Llevant', 'Valencia', 'Madrid', 'Bierzo', 'Uclés','Priorato', 'Bullas',
                         'Cigales','Tarragona','Costers del Segre', 'Méntrida', 'Conca de Barberà',
                         'Manchuela', 'Valdeorras', 'Ribera del Júcar', 'Rueda','Ribera del Guadiana',
                         'Tierra de León', 'Pla de Bages','Arlanza','Alella', 'Sierras de Málaga',
                         'Ribeira Sacra','Binissalem-Mallorca','Monterrei', 'Ribeiro','Tierra del Vino de Zamora',
                         'Valle de la Orotava','Arribes','Málaga','Tacoronte-Acentejo', 'Bizkaiko Txakolina',
                         'Rías Baixas','Ycoden-Daute-Isora', 'Lanzarote','Condado de Huelva','La Palma',
                         'Montilla-Moriles','Getariako Txakolina','Arabako Txakolina', 'Manzanilla',
                         'Valle de Güímar', 'Gran Canaria','Jerez-Xérès-Sherry', 'Cava', 'Abona', 'El Hierrro',
                         'Mondéjar', 'La Gomera',]

vinos_pago = ['Dehesa del Carrizal','Campo de la Guardia','El Terrerazo','Dominio de Valdepusa',
              'Pago Calzadilla','Pago Florentino','Arínzano','Otazu','Prado de Irache','Pago Aylés',
              'Casa del Blanco','Guijoso', 'Finca Élez','Los Balagueses']

vinos_calidad = ['Sierra de Salamanca','Granada Sur-Oeste','Islas Canarias', 'Cangas','Valles de Benavente',
                 'Valtiendas','Lebrija']
```

```python
### Función que asigna denominación a cada región
def denominacion(x):
    if x in denominaciones_origen:
        return "Denominación de Origen"
    
    elif x in vinos_pago:
        return "Vino de Pago"

    elif x in vinos_calidad:
        return "Vino de Calidad"

    else:
        return "Sin denominación"


vinos_cat["Sello_calidad"] = vinos_cat["Región"].apply(denominacion)
vinos_cat.info()
```

```python
'''Ahora que tenemos los datos completos y limpios lo pasamos a csv para poder importarlo''' 

vinos_cat.to_csv("Data/vinos_ESP.csv", index=False)
```
